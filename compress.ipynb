```python
import skvideo.io
import numpy as np
import scipy as sp

video = np.array(skvideo.io.vread("simple_shape.png"))

print(video.shape)
```

```python
print(video)
```

```python
newvideo = np.zeros((3, 16, 16, 3))

#LPF by averaging in order to prevent aliasing

for f in range(3):
    i = 0
    j = 0
    while i < 240:
        while j < 240:
            color0tot = 0
            color1tot = 0
            color2tot = 0
            for k in range(15):
                for l in range(15):
                    color0tot += video[f][i+l][j+k][0]
                    color1tot += video[f][i+l][j+k][1]
                    color1tot += video[f][i+l][j+k][2]
            newvideo[f][i//15][j//15][0] = color0tot // 225.0
            newvideo[f][i//15][j//15][1] = color1tot // 225.0
            newvideo[f][i//15][j//15][2] = color2tot // 225.0
            print((newvideo[f][i//15][j//15][0], newvideo[f][i//15][j//15][1], newvideo[f][i//15][j//15][1]))
            j += 15
        i += 15
            
            
```
