```python
import skvideo.io  
import scipy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
from Helper_functions import *
og_video= skvideo.io.vread("Rolling.png")

frames = og_video.shape[0]
height = og_video.shape[1]
width = og_video.shape[2]
channels = og_video.shape[3]
```

```python
X = og_video.flatten().reshape(frames, height*width*channels)
```

```python
plt.imshow(X.reshape(og_video.shape)[1,:,:],cmap ='gray')
```

```python
pca = PCA(n_components=11)
lower_dimension_data = pca.fit_transform(X)
```

```python
lower_dimension_data.shape
```

```python
approximation = pca.inverse_transform(lower_dimension_data).astype(np.uint8)
approximation.shape
```

```python
plt.imshow(approximation.reshape(og_video.shape)[6,:,:],cmap ='gray')
```
